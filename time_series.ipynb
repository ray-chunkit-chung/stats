```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.stats.api as sm

from statsmodels.tsa.seasonal import seasonal_decompose

%matplotlib inline
```

```python
xs = np.linspace(0,1,1001)
trend = np.linspace(0,10,1001)
noise = np.random.normal(0,0.1,1001)
seasonality = np.sin(xs*np.pi*2)**2

date = pd.Series(['{} days'.format(i) for i in range(1001)]).map(pd.to_timedelta)

series = pd.DataFrame(trend + noise + seasonality + 1, index=date)

# result = seasonal_decompose(series, model='multiplicative')
result = seasonal_decompose(series, model='addictive')
fig = result.plot()
```
